```python
from load_csv import load
# import matplotlib.pyplot as plt

df = load("population_total.csv")
```

```python
df.set_index("country", inplace=True)
```

```python
df
```

```python
morocco_vs_france = df.loc[['Morocco', 'France'], '1800': '2050']
# convert
```

```python
morocco_vs_france
```

```python
def convert(val):
    multiplier = val[-1].lower()
    if multiplier == 'm':
        return float(val[:-1]) * 1000000
    elif multiplier == 'k':
        return float(val[:-1]) * 1000
    else:
        return float(val)

m_vs_f_converted = morocco_vs_france.map(convert)
```

```python
m_vs_f_converted
```

```python
ax = m_vs_f_converted.T.plot(title='Population Projections', xlabel="Year", ylabel='Population')
ax.set_yticks([20000000, 40000000, 60000000], labels=['20M', '40M', '60M'])
```
